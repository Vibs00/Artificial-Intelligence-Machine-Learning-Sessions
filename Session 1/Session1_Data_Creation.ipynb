```python
#Company A's total products data generation
np.random.seed(123)
companyA_total ={
    'product_id': np.random.randint(20500, high =25000, size =200),
}
```

```python
#Company B's total products data generation
np.random.seed(127)
companyB_total ={
    'product_id': np.random.randint(110000, high =120000, size =400),
}
```

```python
#Company A's complaint data generation
np.random.seed(123)
companyA_complaints ={
    'product_id': np.random.randint(19000, high =20500, size =5000),
    'complaint_id': np.random.randint(0, high =100000, size =5000)
}
x =list(set(companyA_complaints['product_id']))
companyA_total['product_id'] =np.append(companyA_total['product_id'], x)
x =x[0:int(len(x)/1.1)]
companyA_complaints['product_id'] =np.append(companyA_complaints['product_id'], x, axis =0)
companyA_complaints['complaint_id'] =np.append(companyA_complaints['complaint_id'], np.random.randint(100000, high =1000000, size =len(x)))
```

```python
#Company B's complaint data generation
np.random.seed(127)
companyB_complaints ={
    'product_id': np.random.randint(100000, high =101500, size =1100),
    'complaint_id': np.random.randint(0, high =100000, size =1100)
}
x =list(set(companyB_complaints['product_id']))
companyB_total['product_id'] =np.append(companyB_total['product_id'], x)
x =x[0:int(len(x)/1.2)]
for _ in range(0, 3):
    x =np.append(x, x)
companyB_complaints['product_id'] =np.append(companyB_complaints['product_id'], x, axis =0)
companyB_complaints['complaint_id'] =np.append(companyB_complaints['complaint_id'], np.random.randint(100000, high =1000000, size =len(x)))
```

```python
#converting to pandas DataFrame
df_companyA_complaints =pd.DataFrame(companyA_complaints)
df_companyB_complaints =pd.DataFrame(companyB_complaints)
```
